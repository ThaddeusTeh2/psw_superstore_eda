#### Bar Chart of Sales by Category (Numbers)

```python
# Group sales by category
sales_by_category = df.groupby("Category")["Sales"].sum().sort_values()

# Plot
plt.figure(figsize=(8,6))
sales_by_category.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Total Sales by Category", fontsize=16, fontweight="bold", pad=15)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Total Sales ($)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Technology has the highest sales among all categories

# Furniture is the second highest

# Office Supplies has the lowest sales
```

#### Pie chart of customer segments


```python
# Count number of records per Segment
segment_counts = df["Segment"].value_counts()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(
    segment_counts,
    labels=segment_counts.index,
    autopct="%1.1f%%",    
    startangle=120,          # rotate for better layout
)
plt.title("Customer Segments Distribution", fontsize=14, fontweight="bold")
plt.show()

# Consumer segment has the highest sales

# Corporate comes next

# Home Office has the lowest sales
```

#### Pie chart of ship mode

```python
# Count number of records per Segment
segment_counts = df["Ship Mode"].value_counts()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(
    segment_counts,
    labels=segment_counts.index,
    autopct="%1.1f%%",      
    startangle=120,        
)
plt.title("Ship Mode Distribution", fontsize=14, fontweight="bold")
plt.show()

# Standard Class is the most used shipping mode

# Second Class and First Class are used less

# Same Day shipping is the least used
```

#### Pie chart of sales by category (Percentage)

```python
# Aggregate total sales by Category
sales_by_category = df.groupby("Category")["Sales"].sum()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(
    sales_by_category,
    labels=sales_by_category.index,
    autopct="%1.1f%%",       
    startangle=120,       
)

plt.title("Total Sales by Category", fontsize=14, fontweight="bold")
plt.show()

# Technology has the highest sales

# Furniture is the second highest

# Office Supplies has the lowest sales
```

#### Grouped bar chart of orders by category and sub-category

```python
df = pd.read_csv("superstore_feature_engineered.csv", encoding='ISO-8859-1')
```

```python
# Count number of orders for each Category + Sub-Category
subcat_counts = df.groupby(["Category", "Sub-Category"]).size().reset_index(name="Order Count")

# Plot
plt.figure(figsize=(12,6))
for category in subcat_counts["Category"].unique():
    subset = subcat_counts[subcat_counts["Category"] == category]
    plt.bar(subset["Sub-Category"], subset["Order Count"], label=category)

plt.title("Orders by Category and Sub-Category", fontsize=16, fontweight="bold")
plt.xlabel("Sub-Category", fontsize=12)
plt.ylabel("Number of Orders", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.legend(title="Category")
plt.tight_layout()
plt.show()

# Each main category is broken down into sub-categories

# Office Supplies has the highest number of orders overall 

# Technology and Furniture have fewer orders compared to Office Supplies
```

# PROFIT VS CATEGORIES/SEGMENTS

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Category", y="Profit", estimator=sum, ci=None)
plt.title("Total Profit by Category")
plt.show()

plt.figure(figsize=(12,6))
sns.barplot(data=df, x="Sub-Category", y="Profit", estimator=sum, ci=None, order=df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False).index)
plt.xticks(rotation=45)
plt.title("Total Profit by Sub-Category")
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Segment", y="Profit", estimator=sum, ci=None)
plt.title("Total Profit by Segment")
plt.show()

# Technology gives the highest profit

# Office Supplies also gives good profit

# Furniture has the lowest profit and sometimes losses
```

# PROFIT VS REGIONS/STATES

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Region", y="Profit", estimator=sum, ci=None)
plt.title("Total Profit by Region")
plt.show()

plt.figure(figsize=(12,6))
top_states = df.groupby("State")["Profit"].sum().sort_values(ascending=False).head(15).index
sns.barplot(data=df[df["State"].isin(top_states)], x="State", y="Profit", estimator=sum, ci=None, order=top_states)
plt.xticks(rotation=45)
plt.title("Top 15 States by Total Profit")
plt.show()

plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="Region_Category", y="Profit")
plt.xticks(rotation=45)
plt.title("Profit Distribution by Region-Category Combination")
plt.show()

# West region makes the highest profit

# East also gives good profit

# South has lower profit

# Central is the least profitable and sometimes shows losses
```

# PROFIT VS DISCOUNT


```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Discount", y="Profit", hue="Category", alpha=0.6)
plt.title("Discount vs Profit (colored by Category)")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Has_Discount", y="Profit")
plt.title("Profit Distribution: With vs Without Discount")
plt.show()

plt.figure(figsize=(8,5))
sns.lineplot(data=df, x="Discount_x_Quantity", y="Profit", ci=None)
plt.title("Profit vs Discount * Quantity")
plt.show()

# When discount is low, profit is mostly positive

# As discount increases, profit goes down

# Very high discounts often lead to losses
```

# PROFIT VS SALES / QUANTITY


```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Sales", y="Profit", alpha=0.6)
plt.xscale("log")
plt.yscale("symlog")
plt.title("Sales vs Profit (Log-Scaled)")
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Quantity", y="Profit", alpha=0.6)
plt.title("Quantity vs Profit")
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Price_Per_Unit", y="Profit_Per_Unit", alpha=0.6)
plt.title("Price per Unit vs Profit per Unit")
plt.show()

# When sales or quantity are small, profit is also small but mostly positive

# With medium sales or quantity, profit can be higher

# At very high sales or very large quantities, many orders still show losses because of big discounts
```

# PROFIT OVER TIME


```python
plt.figure(figsize=(8,5))
sns.lineplot(data=df.groupby("Order_Year")["Profit"].sum().reset_index(), x="Order_Year", y="Profit")
plt.title("Total Profit by Year")
plt.show()

plt.figure(figsize=(10,5))
sns.lineplot(data=df.groupby("Order_Month")["Profit"].sum().reset_index(), x="Order_Month", y="Profit")
plt.title("Total Profit by Month (All Years Combined)")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Order_Quarter", y="Profit")
plt.title("Profit Distribution by Quarter")
plt.show()

plt.figure(figsize=(10,5))
sns.boxplot(data=df, x="Order_DayOfWeek_Name", y="Profit", order=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"])
plt.title("Profit by Day of Week")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Is_Holiday_Season", y="Profit")
plt.title("Profit Distribution: Holiday Season vs Non-Holiday")
plt.show()

# Profit increased steadily from 2014 to 2017

# Within each year, profit goes up and down by month

# December has the highest profit, while some months like April and July are lower

# Overall: Profit is growing year by year, with clear seasonal peaks in December
```

# PROFITABILITY METRICS


```python
plt.figure(figsize=(8,5))
sns.histplot(df["Profit_Margin"], bins=30, kde=True)
plt.title("Distribution of Profit Margins")
plt.show()

plt.figure(figsize=(10,6))
sns.violinplot(data=df, x="Category", y="Profit_Margin")
plt.title("Profit Margin by Category")
plt.show()

# Most orders have positive profit margins between 0 and 0.5

# Some orders show negative margins (losses), with a few very large losses

# By category:
# Technology â†’ generally higher margins but more variation
# Office Supplies â†’ margins are stable but smaller
# Furniture â†’ widest spread, with many orders in losses
```

# CORRELATION HEATMAP


```python
plt.figure(figsize=(10,8))
corr = df[["Profit","Sales","Quantity","Discount","Profit_Margin","Profit_Per_Unit","Sales_Per_Quantity"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap of Key Metrics")
plt.show()

# Discount strongly reduces Profit Margin

# Sales and quantity alone donâ€™t guarantee profit

# Profitability is best explained by discount and margins, not just sales volume
```
